```python
#Importing the required packages
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Dropout,Flatten,MaxPooling2D
import matplotlib.pyplot as plt
import numpy as np


#Explaination:
#Tensorflow:Main deep learning library we are using.it provides tools for building,training, and deploying machine learning model.
#keras.models.sequential:Sequentil is the typ of keras model where layers are added one after the other.
#keras.layers:we import the several type of the layers to define our CNN rchitecture.
#Dense:A fully connectted layer.
#Conv2D:A 2D convolutional layers for applying the convolutional filters to an image.
#Dropout:A regularization layer to reduce overfitting by randomaly setting a fraction of inputs to zero.
#Flatten:A layer to flatten the 2D matrix of features into 1D vactor for fully connecttted layers.
#MaxPooling:A pooling layer to down-sample input, reducing its spatial dimensions.

#matplotlib.pyplot:Used for visualizing images and plotting graphs, useful for monitoring training progress and evaluating the model performance.
#numpy:handling array and numerical operations,used for data manipulation
```

```python
#Loading and preproceessing the image data

mnist = tf.keras.datasets.mnist 
#import the dataset,mnist is variable.
(x_train, y_train),(x_test,y_test) = mnist.load_data() #splitting into training and testing data
input_shape=[28,28,1] #Define the input shape
#28:The height of each image
#28:The width of each images.
#1:The number of the color channels


#Here we loading the MNIST Dataset which contain grayscale images of handwritten digits(0-9) with each image size 28x28 pixels.
#mnist.load_data():Function downloads the mnist dataset.
#(x_train,y_train):x_train contins the images and y_train contains the labels.
#(x_test,y_test):Testing data used for evaluating the model after training.


```

```python
#Making sure that the values are float so that we can get decimal points after division
#Preprocessing the data(Normalizing and REshaping

x_train = x_train.reshape(x_train.shape[0],28,28,1)
x_test = x_test.reshape(x_test.shape[0],28,28,1)

print("Data type of x_train:",x_train.dtype)

#Reshaping:The reshap function adds a fourth dimension to teh arrays,matching the expected input shape [28,28,1]
#It reshape the (10000,28,28) to (10000,28,28,1)

#x_test.shape[0]: This returns the number of images in x_test, which is 10,000.
#28, 28: These values represent the image dimensions, which stay the same.
```

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
print("Data type after converting to float:",x_train.dtype)

#Here we convert the float datatype bacause it is more suiable for the neural network models.
```

```python
#Normalize the RGB codes by dividing it to the max RGB value
#normalize the training and testing data.

x_train = x_train/255
x_test = x_test/255
#now x_test and x_train contain the values between 0 and 1 which is optimal for training deep learning models.

print("Shape of Training:",x_train.shape)
print("shape of training:",x_test.shape)




#"""Pixel values in x_train and x_test are initially in the range [0, 255],
#representing grayscale intensity (0 being black, 255 being white, 
#and values in between representing various shades of gray)."""

#60000: The number of training images.
#28: The height of each image (28 pixels).
#28: The width of each image (28 pixels).
#1: The number of color channels, where 1 indicates grayscale.
#10000: The number of testing images.
```

```python
#Defining the model architecture

model = Sequential()
#initialize the sequentia model.It allow to build the neurl network layer by layerwhere each layer has exactly one input tensor and one output tensor.

model.add(Conv2D(28, kernel_size=(3,3), input_shape=input_shape))
#28 filters: The layer has 28 filters, which means it will learn 28 different feature maps.
#kernel_size=(3,3): Each filter (or kernel) has a size of 3x3 pixels, meaning it will look at 3x3 patches of the input to detect features like edges and textures.
#Specifying this input shape is required only in the first layer.


model.add(MaxPooling2D(pool_size=(2,2)))
#pool_size=(2,2): This layer downsamples the feature maps generated by the Conv2D layer, reducing their spatial dimensions by taking the maximum value in each 2x2 window.


model.add(Flatten())
#Flatten layer converts the multidimensional input into one dimensional vector,which is typiclly required before passing data into dense(fully connectted layers)

model.add(Dense(200,activation = 'relu'))
#this line adds a fully connected layer with 200 neurons.
#Each neuron in this layer is connectted to every neuron in the previous layer.
#Argument applies the Rectified Linear Unit activation function to each neuron,which outputs max(0,x) for each input x.
#Relu helps the model learn complex patterns while avoiding issues like vanishing gradients.

model.add(Dropout(0,3))
#used to prevent overfitting  by randomly setting a fraction of input units to zero during training.
#which means 30% of neurons in this layer will be “dropped out” during training.


model.add(Dense(10,activation= 'softmax'))
#This line adds another dense layer with 10 neurons.
#applies the softmax activation function, which is commonly used for classification tasks where outputs represent probabilities of each class.

model.summary()
#howing each layer’s name, output shape, and the number of parameters to train in each layer.


#This code defines a simpe convolutional neural network model using teh keras sequential API.
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train, epochs=2)
#""This line configures the model for training. It specifies the optimizer, 
#loss function, and evaluation metric.""
#optimizer='adam':The Adam optimizer is an efficient gradient descent algorithm that adjusts the learning rate adaptively for each parameter.
#""loss='sparse_categorical_crossentropy':

#This is the loss function used to measure the difference between predicted and actual labels during training.""
#sparse_categorical_crossentropy is used when you have integer labels (e.g., 0, 1, 2 for three classes) rather than one-hot encoded labels.
#Accuracy calculates the percentage of correct predictions.
#model.fit(x_train, y_train, epochs=2)
#This line trains the model on the data provided.
#epochs=2:

#This specifies the number of times the model will go through the entire training dataset.
#During each epoch, the model performs forward and backward passes to update its weights according to the loss and optimizer.
```

```python
test_loss, test_acc = model.evaluate(x_test,y_test)
#This line evaluates the model’s performance on a separate test dataset, which the model has not seen before. The goal is to check how well the model generalizes to new data.

print("Loss=%.3f" %test_loss)
#This line prints the test loss to three decimal places.

print("Loss=%.3f" % test_loss)
#Loss=%.3f" % test_loss:%.3f is a format specifier that rounds the loss to three decimal places for a more readable result.


print("Accuracy=%.3f" %test_acc)
#Accuracy=%.3f" % test_acc:Similarly, %.3f is used to round the accuracy to three decimal places.
#print("Accuracy=%.3f" % test_acc):This line prints the test accuracy to three decimal places.
```

```python
#Showing the image at positional[] from datasets
image = x_train[0]
#image = x_train[0]:This line selects the first image from the x_train dataset and assigns it to the variable image
#x_train[0] accesses the first element of the


plt.imshow(np.squeeze(image),cmap='gray')
#plt.inp.squeeze(image):
#np.squeeze() removes any single-dimensional entries from the shape of an array.mshow(np.squeeze(image), cmap='gray'):This line displays the image using the imshow function from matplotlib.pyplot.

plt.show()
```

```python
#Predicating the class of the image
image = image.reshape(1,image.shape[0],image.shape[1],image.shape[2])
#This line reshapes the image array to add a new dimension at the beginning.
#image.shape[0], image.shape[1], and image.shape[2] represent the original height, width, and depth (or channels) of the image.

predict_model = model.predict([image])
#predict_model = model.predict([image]):This line uses the model to make a prediction on the reshaped image.
#model.predict([image]): model.predict() takes an array of images (even if there’s only one image in this case) and returns the predicted probabilities for each class.
#
print("Predicted class: {}".format(np.argmax(predict_model)))
#This line prints out the predicted class based on the model’s output.
#np.argmax() finds the index of the highest probability in predict_model, which corresponds to the predicted class.
#The format method inserts the predicted class index into the string to display it.
```
